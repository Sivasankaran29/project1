```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time

# Initialize the WebDriver
driver = webdriver.Chrome()
driver.implicitly_wait(5)

try:
    # Open the initial URL
    driver.get('https://www.redbus.in/')
    driver.maximize_window()
    time.sleep(5)

    # Scroll down to load more content
    for _ in range(20):
        body_element = driver.find_element(By.TAG_NAME, 'body')
        body_element.send_keys(Keys.ARROW_DOWN)
        time.sleep(0.5)

    try:
        right = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="Carousel"]/span/span/i'))
        )
        for _ in range(4):  # Click 6 times to move to the desired position
            right.click()
            time.sleep(1)  # Small delay between clicks
            print("Clicked on right")
    except Exception as e:
        print(f"Failed to click on right: {e}")
        driver.quit()
        exit()

    # Click on the image to select WBTC
    try:
        pepus = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="Carousel"]/div[12]/div[1]/div[1]/div[2]/div[1]/div[1]'))
        )
        pepus.click()
        print("Clicked on pepus")
    except Exception as e:
        print(f"Failed to click on pepus: {e}")
        driver.quit()
        exit()

    time.sleep(2)

    # Collect all route links
    route_links = []
    page_number = 1
    while True:
        for _ in range(40):
            body_element = driver.find_element(By.TAG_NAME, 'body')
            body_element.send_keys(Keys.ARROW_DOWN)
            time.sleep(1)
        try:
            # Collect route links on the current page
            route_link_elements = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[2]/div[1]/a'))
            )
            route_links.extend([element.get_attribute('href') for element in route_link_elements])
            print(f"Found {len(route_links)} route links.")
        except Exception as e:
            print(f"No more route links found: {e}")
            break

        try:
            # Collect route links on the current page
            route_link_elements = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[3]/div[1]/a'))
            )
            route_links.extend([element.get_attribute('href') for element in route_link_elements])
            print(f"Found {len(route_links)} route links.")
        except Exception as e:
            print(f"No more route links found: {e}")
            break

        try:
            # Collect route links on the current page
            route_link_elements = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[4]/div[1]/a'))
            )
            route_links.extend([element.get_attribute('href') for element in route_link_elements])
            print(f"Found {len(route_links)} route links.")
        except Exception as e:
            print(f"No more route links found: {e}")
            break

        try:
            # Collect route links on the current page
            route_link_elements = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[5]/div[1]/a'))
            )
            route_links.extend([element.get_attribute('href') for element in route_link_elements])
            print(f"Found {len(route_links)} route links.")
        except Exception as e:
            print(f"No more route links found: {e}")
            break

        try:
            # Collect route links on the current page
            route_link_elements = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[6]/div[1]/a'))
            )
            route_links.extend([element.get_attribute('href') for element in route_link_elements])
            print(f"Found {len(route_links)} route links.")
        except Exception as e:
            print(f"No more route links found: {e}")
            break

        try:
            # Collect route links on the current page
            route_link_elements = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[7]/div[1]/a'))
            )
            route_links.extend([element.get_attribute('href') for element in route_link_elements])
            print(f"Found {len(route_links)} route links.")
        except Exception as e:
            print(f"No more route links found: {e}")
            break

        try:
            # Collect route links on the current page
            route_link_elements = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[8]/div[1]/a'))
            )
            route_links.extend([element.get_attribute('href') for element in route_link_elements])
            print(f"Found {len(route_links)} route links.")
        except Exception as e:
            print(f"No more route links found: {e}")
            break

        try:
            # Collect route links on the current page
            route_link_elements = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[9]/div[1]/a'))
            )
            route_links.extend([element.get_attribute('href') for element in route_link_elements])
            print(f"Found {len(route_links)} route links.")
        except Exception as e:
            print(f"No more route links found: {e}")
            break

        try:
            # Collect route links on the current page
            route_link_elements = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[10]/div[1]/a'))
            )
            route_links.extend([element.get_attribute('href') for element in route_link_elements])
            print(f"Found {len(route_links)} route links.")
        except Exception as e:
            print(f"No more route links found: {e}")
            break

        try:
            # Collect route links on the current page
            route_link_elements = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[11]/div[1]/a'))
            )
            route_links.extend([element.get_attribute('href') for element in route_link_elements])
            print(f"Found {len(route_links)} route links.")
        except Exception as e:
            print(f"No more route links found: {e}")
            break

        try:
            # Locate and click on the element for the next page
            next_page_link = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, f"//*[@id='root']/div/div[4]/div[12]/div[text()='{page_number + 1}']"))
            )
            next_page_link.click()
            # Increment page number for the next iteration
            page_number += 1

        except Exception as e:
            print("no mor page")
            break
        try:
            top = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="c_scrollToTop"]/span'))
            )
            top.click()
            print("Clicked on top")
        except Exception as e:
            print(f"Failed to click on top: {e}")

    time.sleep(2)

    # Print all collected route links
    print(route_links)

    # Iterate through each route link to scrape data
    product_details = []
    for route_link in route_links:
        try:
            driver.get(route_link)
            time.sleep(3)
            # Adjust as needed to ensure page loads completely
            for _ in range(5):
                body_element = driver.find_element(By.TAG_NAME, 'body')
                body_element.send_keys(Keys.ARROW_DOWN)
                time.sleep(1)

            time.sleep(2)
            # Get route name
            try:
                route_name_element = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.CSS_SELECTOR, "h1[class='D136_h1']"))
                )
                route_name = route_name_element.text
            except Exception as e:
                print(f"Failed to get route name for {route_link}: {e}")
                route_name = 'N/A'

            for _ in range(5):
                body_element = driver.find_element(By.TAG_NAME, 'body')
                body_element.send_keys(Keys.ARROW_DOWN)
                time.sleep(1)

            time.sleep(2)
            try:
                SHOWBUS = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH, '//*[@id="result-section"]/div[3]/div/div[2]/div/div[4]/div[2]'))
                )
                SHOWBUS.click()
                print("Clicked on SHOWBUS")
            except Exception as e:
                print("no show bus")

            time.sleep(1)

            try:
                SHOWBUS = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH, '//*[@id="result-section"]/div[2]/div/div[2]/div/div[4]/div[2]'))
                )
                SHOWBUS.click()
                print("Clicked on SHOWBUS")
            except Exception as e:
                print("no show bus")
            time.sleep(1)
            try:
                SHOWBUS = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH, '//*[@id="result-section"]/div[1]/div/div[2]/div/div[4]/div[2]'))
                )
                SHOWBUS.click()
                print("Clicked on SHOWBUS")
            except Exception as e:
                print("no show bus")

            scroll_height = driver.execute_script("return document.body.scrollHeight")
            for i in range(0, scroll_height, 100):
                driver.execute_script(f"window.scrollTo(0, {i});")
                time.sleep(1)
            print("Scrolled to bottom")

            # Collect bus details
            bus_names_elements = driver.find_elements(By.CSS_SELECTOR, "div[class='travels lh-24 f-bold d-color']")
            bus_type_elements = driver.find_elements(By.CSS_SELECTOR, "div[class='bus-type f-12 m-top-16 l-color evBus']")
            start_time_elements = driver.find_elements(By.CSS_SELECTOR, "div[class='dp-time f-19 d-color f-bold']")
            end_time_elements = driver.find_elements(By.CSS_SELECTOR, "div[class='bp-time f-19 d-color disp-Inline']")
            dur_elements = driver.find_elements(By.CSS_SELECTOR, "div[class='column-four p-right-10 w-10 fl']")
            price_elements = driver.find_elements(By.CSS_SELECTOR, "span[class='f-19 f-bold']")
            rating_elements = driver.find_elements(By.CSS_SELECTOR, "div[class='rating-sec lh-24']")
            seat_elements = driver.find_elements(By.CSS_SELECTOR, "div[class='seat-left m-top-16']")

            bus_names = [element.text for element in bus_names_elements]
            bus_type = [element.text for element in bus_type_elements]
            departing_time = [element.text for element in start_time_elements]
            reaching_time = [element.text for element in end_time_elements]
            duration = [element.text for element in dur_elements]
            price = [element.text for element in price_elements]
            rating = [element.text for element in rating_elements]
            seats_available = [element.text for element in seat_elements]

            for i in range(len(bus_names_elements)):
                bus_details = {
                    'busname': bus_names_elements[i].text,
                    'bustype': bus_type_elements[i].text if i < len(bus_type_elements) else 'N/A',
                    'route_name': route_name,
                    'route_link' : route_link,
                    'departing_time': start_time_elements[i].text if i < len(start_time_elements) else 'N/A',
                    'reaching_time': end_time_elements[i].text if i < len(end_time_elements) else 'N/A',
                    'duration': dur_elements[i].text if i < len(dur_elements) else 'N/A',
                    'price': price_elements[i].text if i < len(price_elements) else 'N/A',
                    'rating': rating_elements[i].text if i < len(rating_elements) else 'N/A',
                    'seats_available': seat_elements[i].text if i < len(seat_elements) else 'N/A'
                }
                product_details.append(bus_details)

            print(f"Scraped data for route: {route_name}")

        except Exception as e:
            print(f"Failed to scrape data for route {route_link}: {e}")

    # Output collected data
    for detail in product_details:
        print(detail)

except Exception as e:
    print(f"Error during scraping: {e}")

finally:
    driver.quit()
```

```python
import csv
import pandas as pd
print(product_details)

csv_file = 'pepus.csv'
csv_columns = ['busname', 'bustype', 'route_name', 'route_link' ,'departing_time', 'reaching_time', 'duration', 'price', 'rating', 'seats_available']
try:
    with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in product_details:
            writer.writerow(data)
    print(f"CSV file '{csv_file}' successfully saved.")
except IOError:
    print(f"Failed to write CSV file '{csv_file}'")
pd.read_csv('pepus.csv')
```
